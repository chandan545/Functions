1. What is the difference between a function and a method in Python?
   - Function
     - A block of code that is defined outside any class and can be called on its own.
     - Called independently or by passing arguments directly.
     - Can be used anywhere in the program, provided it is in scope.

           def add(x, y): return x + y

   - Method
     - A function that is defined inside a class and must be called on an object of that class.
     - Called on an object or within a class context.
     - Limited to the objects or class they are associated with.

           class Math: def add(self, x, y): return x + y

   - In Python, functions arguments and parameters are closely related concepts that deal with how values are passed to and handled by functions.
     - Arguments:

3. What are the different ways to define and call a function in Python?
   - In Python, functions can defined and called in various ways depending on their purpose and usage. Below are the different approaches to defining and calling functions, with examples for each:
   1. Regular Function
      - Use the "def" keyword to define a function.
                    def greet(name):
                        return f"Hello, {name}!"
      - Calling:    
                    print(greet("Chandan"))  # Output: Hello, Chandan!
   2. Function with Default Arguments
      - Define parameters with default values to make them optional.
                    def greet(name="Guest"):
                        return f"Hello, {name}!"
      - Calling:
                   print(greet())           # Output: Hello, Guest!
                   print(greet("Chandan"))  # Output: Hello, Chandan!

   3. Lambda (Anonymous) Function
      - Use the "lambda" keyword to create a single-expression function without a name.
                    square = lambda x: x ** 2
      - Calling:
                    print(square(5))  # Output: 25
   4. Function with *args
      - Define a function to accept an arbitrary number of positional arguments.
                   def add_all(*args):
                       return sum(args)
      - Calling:    
                   print(add_all(1, 2, 3, 4))  # Output: 10
   5. Function with **kwargs
      - Define a function to accept an arbitrary number of keyword arguments.
                   def describe_person(**kwargs):
                   return ", ".join(f"{key}: {value}" for key, value in kwargs.items())
      - Calling:
                  print(describe_person(name="Chandan", age=25, city="Haldwani"))
                  # Output: name: Chandan, age: 25, city: Haldwani
   6. Nested Function
      - Define one function inside another.
                  def outer_function(greeting):
                      def inner_function(name):
                          return f"{greeting}, {name}!"
                  return inner_function
      - Calling:
                  hello = outer_function("Hello")
                  print(hello("Chandan"))  # Output: Hello, Chandan!
   7. Recursive Function
      - A function that calls itself to solve a problem iteratively.
                  def factorial(n):
                      if n == 0:
                         return 1
                      return n * factorial(n - 1)
      - Calling:
                  print(factorial(5))  # Output: 120
   8. Generator Function
      - Use the "yield" keyword to return values lazily, creating an iterator.
                  def countdown(n):
                      while n > 0:
                        yield n
                        n -= 1
      - Calling:
                 for num in countdown(5):
                     print(num)
                 # Output: 5 4 3 2 1
   9. Higher-Order Function
      - A function that accepts another function as an argument or returns a function.
                 def apply_func(func, value):
                     return func(value)
      - Calling:
                 print(apply_func(lambda x: x * 2, 5))  # Output: 10
   10. Method in a Class
       - Define a function inside a class (becomes a method).
                 class Greeter:
                    def greet(self, name):
                        return f"Hello, {name}!"
       - Calling:
                 greeter = Greeter()
                 print(greeter.greet("Chandan"))  # Output: Hello, Chandan!
   11. Static Method
       - Define a method that doesn't use the class or instance directly, with @staticmethod.
                  class Math:
                    @staticmethod
                    def add(a, b):
                        return a + b
       - Calling:
                 print(Math.add(3, 5))  # Output: 8
   12. Class Method
       - Define a method that operates on the class itself, with @classmethod.
                  class Counter:
                      count = 0

                     @classmethod
                     def increment(cls):
                         cls.count += 1

5. What are iterators in Python and how do they differ from iterables?
   - Iterables
     - An iterable is an object capable of returning its elements one at a time. Examples include sequences like lists, tuples, strings, and other objects like sets and dictionaries.
     - An iterable has a special method__iter__()that returns an iterator.
     - Iterables can be used in a for loop or with functions like list(), max (), min(), etc.

     Example:
            # Examples of iterables
            my_list = [1, 2, 3]       # List
            my_tuple = (4, 5, 6)      # Tuple
            my_string = "Python"      # String

            # Iterating over an iterable
            for char in my_string:
            print(char)

   - Iterators
     - An iterator is an object that represents a stream of data. It implements the methods:
     -a__iter__(): Returns the iterator object itself.
     -b__next__(): Returns the next value from the iterator. If there are no more items, it raises a StopIteration exception.
    
     Example:
            my_list = [1, 2, 3]       # This is an iterable
            iterator = iter(my_list)  # Create an iterator from the iterable

            # Using the iterator
            print(next(iterator))  # Output: 1
            print(next(iterator))  # Output: 2
            print(next(iterator))  # Output: 3
            # next(iterator)       # Raises StopIteration

6. Explain the concept of generators in Python and how they are defined.
   - In Python, generators are a type of iterable that produce values lazily, generating items one at a time as needed instead of storing them in memory. They are memory-efficient and ideal for handling large or infinite sequences.
   - Generators are defined using functions with the 'yield' keyword. When called, a generator function returns a generator object that can be iterated.

   Example:
        def count_up_to(max):
            count = 1
            while count <= max:
                yield count
                count += 1

  Usage:
        gen = count_up_to(3)
        print(next(gen))  # Output: 1
        print(next(gen))  # Output: 2

    - They can also be created using generator expressions:

          gen = (x ** 2 for x in range(5))

7. What are the advantages of using generators over regular functions?
   - Generators are more memory efficient than regular functions because they produce values one at a time, instead of creating an entire sequence in memory. This makes them ideal for working with large or infinite sequences.
   - Generators can be faster than regular loops, especially when working with large data sets.
   - Generators can make code easier to read and understand by breaking up the iteration process into smaller chunks.
   - Generators follow the principle of lazy evaluation, where values are computed only when needed.
   - In Python, a generator function is similar to a normal function, but it uses the 'yield' keyword instead of 'return'.

9. Explain the purpose and usage of the `map()` function in Python.
   - The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results.
   - Transformation: Used to transform each item in an iterable by applying the specified function.
     - Syntax:
       map(function, iterable)

   - Example:
       
         numbers = [1, 2, 3, 4]
         squared = map(lambda x: x * x, numbers)
         print(list(squared))

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - `map()`:
       - Applies a function to each item in an iterable and returns a new iterable with the results.
       - Transform elements.
  
               nums = [1, 2, 3]
               squares = map(lambda x: x**2, nums)
               print(list(squares))  # Output: [1, 4, 9]

    - `reduce()`(from `functools` module):
       - Reduces an iterable to a single value by repeatedly applying a binary function(two arguments).
       - Aggregate values(e.g., sum, product)
       
              from functools import reduce
              nums = [1, 2, 3, 4]
              result = reduce(lambda x, y: x * y, nums)
              print(result)         #output: 24

    - `filter()`
       - Filter elements of an iterable based on a predicate function.
       - Select elements thaat satisfy a condition.
         
             nums = [1, 2, 3, 4]
             evens = filter(lambda x: x % 2 == 0, nums)
             print(list(evens))    #output: [2, 4]

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
#ans.
def sum_even_numbers(numbers):
    sum = 0
    for num in numbers:
        if num % 2 == 0:
            sum += num
    return sum
```

```python
sum_even_numbers([1, 2, 3, 4, 5])
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
#ans.
a = range(1, 201)
b = filter(lambda c: all(c % i != 0 for i in range(2, int(c**0.5) + 1)), a)
print(list(b))
print("prime")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
#ans.
n = int(input("Enter a specific number:"))
def fib(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        a, b = b, a+b
fib_series = fib(n)
print(list(fib_series))
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
#ans.
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example:
for power in powers_of_2(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
#ans.
file_content = [
    "Hello, this is the first line.",
    "This is the second line.",
    "And this is the third line.",
    "This is the fourth line.",
]


def read_file_line_by_line(content):
    for line in content:
        yield line.strip()


for line in read_file_line_by_line(file_content):
    print(line)
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
#ans.
# List of tuples
data = [(1, 3), (2, 1), (4, 2), (3, 5)]

# Sort the list based on the second element of each tuple using lambda
sorted_data = sorted(data, key=lambda x: x[1])

# Print the sorted list
print(sorted_data)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
#ans.
celsius_temps = [0, 20, 30, 40, 100]
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print(fahrenheit_temps)
```
